```python
from sympy import *
from sympy.plotting import plot3d, plot
import matplotlib.pyplot as plt
import numpy as np

x, y = symbols("x y")
```

```python
def task(f, xlim, ylim):
    sols = solve(Eq(f, 0), (x, y))

    print(sols)

    for sol in sols:
        if x in sol:
            plt.axhline(sol[1], color='g', label=f"y={sol[1]}")
        if y in sol:
            plt.axvline(sol[0], color='r', label=f"x={sol[0]}")

    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.legend()
    plt.xlim((xlim[0], xlim[1]))
    plt.ylim((ylim[0], ylim[1]))

    plot3d(f, (x, xlim[0], xlim[1]), (y, ylim[0], ylim[1]))
```

```python
f = x * (y - 5) ** 2
task(f, (-5, 5), (0, 10))
```

```python
f = (y - E/20) ** 2 * sin(pi * x)
task(f, (-2, 2), (-1, 1))
```

```python
f = x * sin(x) * sin(y - 2)
task(f, (-2*np.pi, 2*np.pi), (-2*np.pi, 2*np.pi))
```
