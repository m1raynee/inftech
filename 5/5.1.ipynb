	f(x,y)=e^(-0.9x^2-0.45(x-y)^2 ). Задайте промежутки для икс и игрек (x,-3,3), (y,-3,3). Это делается точно также, как и для случая одной переменной, только в функцию plot3d вместо одного кортежа нужно передать два. 
	f(x,y)=x cos(y)
	Функция f зависит от оценок (по пятибалльной шкале) трех членов экспертной группы и возвращает вероятность принятия положительного решения.
	Функция f зависит от вероятности того, что пойдет дождь, температуры за окном и оценки по десятибалльной шкале заинтересованности данного человека некоторым мероприятием. Возвращает значение 0, если человек не пойдет на это мероприятие, и 1, если пойдет. Температура может варьироваться от -30 до 35.



```python
from sympy import *
from sympy.plotting import plot3d

x, y = symbols('x y')
```

```python
f = sin(sqrt(x ** 2 + y ** 2))
plot3d(f)
```

```python
f = exp(-0.9*x**2 - 0.45*(x-y)**2)
plot3d(f, (x, -3, 3), (y, -3, 3))
```

```python
f = x * cos(y)
plot3d(f)
```

```python
x1, x2, x3 = symbols('x1 x2 x3')
f = ((x1+x2+x3)/3 - 1)/4
f = f.subs(x3, 2)
plot3d(f, (x1, 1, 5), (x2, 1, 5), 
       title = "Вероятность (оценка эксперта 3 = 2)")
```

```python
rain, temp, interest = symbols("rain, temp, interest")

temp_norm = (temp + 30) / 65
interest_norm = interest / 10
fac = interest_norm * (1 - rain) * temp_norm
f = Heaviside(fac - 0.4)
f = f.subs(temp, 24)
plot3d(f, (rain, 0, 1), (interest, 0, 10),
       xlabel = "rain",
       ylabel="interest",
       title="При temp = 24")
```
