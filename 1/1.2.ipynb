**1** Постойте графики функций: квадратичной, кубической, экспоненциальной, показательной. (Коэффициенты вы выбираете самостоятельно)

```python
from sympy import *
from sympy.plotting import plot

x = Symbol('x')
```

```python
f = - E ** x + 1000
```

```python
if all(f.subs(x, -i/10) == f.subs(x, i/10) for i in range(-100, 100)):
    print('Четная')
elif all(f.subs(x, -i/10) == -f.subs(x, i/10) for i in range(-100, 100)):
    print('Нечетная')
else:
    print('Общего вида')

plot(f)
```
